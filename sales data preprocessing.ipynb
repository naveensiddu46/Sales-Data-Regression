# Importing required Liabraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the dataset

```python
sales=pd.read_csv('C:\\Users\\Naveen\\Downloads\\sales.csv',encoding="cp1252")
sales.head()
```

Head function is used to display the top five columns of the data

# Understanding the Dataset

```python
sales.info()
```

To Find which kind of data type like int,float,object etc

# checking the shape of the data

```python
sales.shape
```

shape is used to find the number of observations i.e, Number of Rows and columns

# Finding the Duplicate Values:

```python
sales.duplicated().sum()
```

Duplicated values are nothing but repeated values in your data

# Finding the Null values:

```python
sales.isnull().sum()
```

# Filling of Null values:

```python
sale=sales.fillna(method='bfill')
sale.head()
```

Labelencoder is used to convert the catagorical data into Numeral data.

# Checking wheather Null values Filled or not

```python
sale.isnull().sum()
```

# STATUS VS SALES:

```python
sns.catplot(x='STATUS',y='SALES',data=sale,kind='bar',height=10,palette='muted',ci=5)
```

sales are high for disputed products

# MONTH_ID VS SALES:

```python
sns.catplot(x='MONTH_ID',y='SALES',data=sale,kind='bar',height=15,palette='cividis',ci=5)
plt.show()
```

Sales are high in the month of April

And low in the month of January

# YEAR_ID VS SALES:

```python
sns.catplot(x='YEAR_ID',y='SALES',data=sale,kind='bar',height=10,palette='viridis',ci=5)
plt.show()
```

sales are increasing from last 2 years

# PRODUCTLINE VS SALES

```python
sns.catplot(x='PRODUCTLINE',y='SALES',data=sale,kind='bar',height=10,palette='viridis',ci=5)
plt.show()
```

classic cars are selling more

# COUNTRY VS SALES:

```python
plt.figure(figsize=(12,10))
sns.catplot(x='COUNTRY',y='SALES',data=sale,kind='bar',height=10,palette='viridis',ci=5)
plt.xticks(rotation=65)
plt.show()
```

# DEALSIZE VS SALES:

```python
plt.figure(figsize=(12,10))
sns.catplot(x='DEALSIZE',y='SALES',data=sale,kind='bar',height=10,palette='viridis',ci=5)
plt.xticks(rotation=60);
```

# Dropping Unnecessary Columns:

```python
sa=sale.drop(['ORDERNUMBER','ORDERLINENUMBER','ORDERDATE','QTR_ID','ADDRESSLINE2','PRODUCTCODE','CUSTOMERNAME','PHONE','ADDRESSLINE1','ADDRESSLINE2','CITY','STATE','POSTALCODE','TERRITORY','CONTACTLASTNAME','CONTACTFIRSTNAME'],axis=1)
sa.head()
```

# Pair plot:

```python
sns.pairplot(sa)
```

# Heatmap:

```python
plt.figure(figsize=(10,12))
sns.heatmap(sa.corr(),annot=True)
```

From the above plot we can see  MSRP,PRICEEACH,QUANTITYORDERED are effecting more

# Boxplot:

```python
plt.figure(figsize=(12,10))
plt.xticks(rotation=65)
sns.boxplot(data=sa)
plt.show()
```

boxplot is used to findout the outliers in your data
